```python

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import plotly.graph_objects as go
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
class repair_cost_predictor:
  def __init__(self, xdata, ydata):
    self.xdata = xdata
    self.ydata = ydata

  def train_test_split(self, test_size):
    self.test_size = test_size
    self.data_split = train_test_split(self.xdata, self.ydata, test_size = self.test_size, random_state = 42)

  def model_train(self, model_type):
    self.model_type = model_type
    self.model_trained = self.model_type.fit(self.data_split[0], self.data_split[2])

  def model_predict(self):
   self.model_prediction_for_xtrain = self.model_trained.predict(self.data_split[1])

  def model_accuracy(self, accuracy_messure_method1, accuracy_messure_method2):
    self.accuracy_messure_method1 = accuracy_messure_method1
    self.accuracy_messure_method2 = accuracy_messure_method2
    self.accuracy1 = self.accuracy_messure_method1(self.data_split[3], self.model_prediction_for_xtrain)
    self.accuracy2 = self.accuracy_messure_method2(self.data_split[3], self.model_prediction_for_xtrain)
    print(f"{self.accuracy_messure_method1.__name__} score for this model is {self.accuracy1}")
    print(f"{self.accuracy_messure_method2.__name__} score for this model is {self.accuracy2}")
    return self.accuracy1, self.accuracy2

  def model_presition_plot(self):
        # Crear un DataFrame para los datos
    self.df = pd.DataFrame({'Equipo Dañado': self.data_split[1].flatten(), 'Costo de Reparación': self.data_split[3].flatten(), 'Predicción': self.model_prediction_for_xtrain.flatten()})

    # Graficar con Plotly
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=self.df['Equipo Dañado'], y=self.df['Costo de Reparación'], mode='markers', name='Costo de Reparación', marker=dict(color='green')))
    fig.add_trace(go.Scatter(x=self.df['Equipo Dañado'], y=self.df['Predicción'], mode='lines', name='Regresión lineal', line=dict(color='black')))
    fig.update_layout(title='Regresión Lineal - Costo de Reparación vs Equipo Dañado')
    fig.show()


xdata = 2 * np.random.rand(1000, 1)
ydata = 4+3 * xdata + np.random.rand(1000, 1)

data = repair_cost_predictor(xdata, ydata)
data.train_test_split(0.25)
data.model_train(LinearRegression())
data.model_predict()
data.model_accuracy(mean_squared_error, r2_score);
data.model_presition_plot()

```
